```python
import numpy as np
import matplotlib.pyplot as plt
```

## ReLU (Rectified Linear Unit) activation function

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/bb2c32931fad595832c8e66f2f73760ebcbc0096)

```python
def Relu(x): 
    y=[]
    for i in x:
        if i>=0:
            y.append(i)
        else:
            y.append(0)
    return y
    
```

```python
x=np.array(range(-10,11,1))
plt.plot(x,Relu(x), linewidth='4')
plt.plot(x,x, color='red')
```

## Leaky ReLU
a small, positive gradient when the unit is not active

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/b164641d3e0255fe37afe613a2464f85d83dfe6e)

```python
def LRelu(x): 
    y=[]
    for i in x:
        if i>0:
            y.append(i)
        else:
            y.append(0.2*i)
    return y
    
```

```python
x=np.array(range(-10,11,1))
plt.plot(x,LRelu(x), linewidth='4')
plt.plot(x,x, color='red')
```

```python
x=np.array(range(-10,11,1))
plt.plot(x,Relu(x), linewidth='4', color='y')
plt.plot(x,LRelu(x), linewidth='4', color='black')
plt.plot(x,x, color='red')
```
