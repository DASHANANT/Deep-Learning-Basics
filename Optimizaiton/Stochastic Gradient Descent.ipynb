```python
import numpy as np
import matplotlib.pyplot as plt
```

# Stochastic gradient descent (SGD)
is an iterative method for optimizing an objective function with suitable smoothness properties 

## Point Update Function

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/dda3670f8a8996a0d3bf80856bb4a166cc8db6d4)

```python
def gradient_descent(gradient, start, alpha, n_iter=50, tolerance=1e-06):
    vector = start
    l=[]
    l.append(vector)
    for _ in range(n_iter):
        diff = -alpha * gradient(vector)
        if np.all(np.abs(diff) <= tolerance):            
            break
        vector += diff
        l.append(vector)
    return l
```

```python
L=gradient_descent(gradient=lambda v: 2 * v, start=10.0, alpha=0.2)
print('Minimizer is' ,L[-1])
K=gradient_descent(gradient=lambda v: 2 * v, start=10.0, alpha=0.8)
print('Minimizer is' ,K[-1])
```

```python
x=np.array(range(-10,11,1))
plt.plot(x,x**2)
L=np.array(L)
K=np.array(K)
plt.scatter(L,L**2,color='red')
plt.plot(L,L**2,color='r')
plt.scatter(K,K**2,color='b')
plt.plot(K,K**2,color='b')
```

### Conclusion - learning rate plays major role

```python
K=gradient_descent(gradient=lambda x: (1/3)*x**4-x+9, start=10.0, alpha=0.0005)
print('Minimizer is' ,K[-1])
K=np.array(K)
x=np.array(range(-10,10,1))
plt.plot(x, (1/3)*x**4-x+9)
plt.scatter(K,(1/3)*K**4-K+9,color='r')
plt.plot(K,(1/3)*K**4-K+9,color='r')
```
